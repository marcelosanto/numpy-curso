```python
import pandas as pd
```

```python
dados = {
    'Times': ['Flamengo', 'Palmeiras', 'Vasco', 'Fluminense', 'Santos'],
    'Partidas_Jogadas': [10,10,10,10,10],
    'Vitórias': [7,5,6,4,3],
    'Empates': [5,2,3,4,5],
    'Derrotas': [1,2,3,3,4],
    'Gols_Pro': [22,18,16,15,13],
    'Gols_Contra': [10,12,15,16,20]
}
```

```python
df = pd.DataFrame(dados)
df
```

```python
df['Saldo_Gols'] = df['Gols_Pro'] - df['Gols_Contra']
df
```

```python
df['Pontos'] = df['Vitórias'] * 3 + df['Empates'] + 1
df
```

```python
df = df.sort_values(by='Pontos', ascending=False)
df
```

```python
df.to_excel('../dados/campeonato_futebol2.xlsx', index=False)
```
